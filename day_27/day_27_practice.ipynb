# Day 27 — Regex & Python Functions Practice

```python
import re

rows = []
with open("messy_data.csv", "r") as f:
    next(f)
    for line in f:
        rows.append(line.strip().strip('"'))

rows[:5]
```

## Exercise 1 — Clean each row using regex

```python
def clean_row(row):
    parts = row.split(",")

    name = parts[0].strip()
    city = re.sub(r'[^A-Za-z]', '', parts[1])
    price = int(re.sub(r'\D', '', parts[2]))
    order_id = re.findall(r'ORD\d+', parts[3])[0]

    return name, city, price, order_id


cleaned = [clean_row(r) for r in rows]
print(cleaned)


```

## Exercise 2 — Use map()

```python
prices = list(map(lambda r: int(re.sub(r'\D', '', r.split(",")[2])), rows))
prices
```

## Exercise 3 — Use filter()

```python
expensive = list(filter(lambda r: int(re.sub(r'\D', '', r.split(",")[2])) > 1200, rows))
expensive
```

## Exercise 4 — Use set()

```python
cities = set(map(lambda r: re.sub(r'[^A-Za-z]', '', r.split(",")[1]), rows))
cities
```
