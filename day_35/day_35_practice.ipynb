# Day 35 — Final Data Cleaning Before EDA (Practice)

```python

import pandas as pd
import numpy as np

df = pd.read_csv("day_35_learning_platform_dirty.csv")
df.head()
```

## Exercise 1 — Inspect inconsistent categories

```python
# TODO

df["course_category"].value_counts()
df["subscription_type"].value_counts()
```

## Exercise 2 — Standardize course_category

```python
# TODO

df["course_category"] = (
    df["course_category"]
    .str.strip()
    .str.lower()
)

category_map = {
    "data science": "data science",
    "ds": "data science",
    "web dev": "web development",
    "web development": "web development",
    "ai": "ai"
}

df["course_category"] = df["course_category"].replace(category_map)
```

## Exercise 3 — Handle skewed & invalid values

```python
# TODO

df.loc[(df["age"] < 18) | (df["age"]>30), "age"] = np.nan
df.loc[(df["completion_rate"] < 0) | (df["completion_rate"] > 1), "completion_rate"] = np.nan
```

## Exercise 4 — Conditional cleaning rules

```python
# TODO

df.loc[df["hours_spent"] == 0, "completion_rate"] = 0
```

## Exercise 5 — Create final analysis-ready dataset

```python
# TODO

final_cols = [
    "user_id",
    "course_category",
    "subscription_type",
    "age",
    "hours_spent",
    "completion_rate"
]

final_df = df[final_cols]
final_df.head()
```
