## Exercise 2 — Business-key duplicates

```python
# TODO

df[df.duplicated(subset = ["device_id", "event_type", "event_time"])]
```

## Exercise 3 — Remove duplicates

```python
# TODO

df_clean = df.drop_duplicates(
    subset = ["device_id", "event_type", "event_time"],
    keep = "last"
)

df_clean
```

## Exercise 4 — Parse dates

```python
# TODO

df_clean["event_time"] = pd.to_datetime(
    df_clean["event_time"],
    errors = "coerce"
)

df_clean
```

## Exercise 5 — Extract date parts

```python
# TODO

df_clean["event_date"] = df_clean["event_time"].dt.date
df_clean["event_hour"] = df_clean["event_time"].dt.hour

df_clean
```
