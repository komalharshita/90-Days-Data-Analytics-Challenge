# Day 30 — Outlier Detection & Data Type Optimization

```python

import pandas as pd
import numpy as np

df = pd.read_csv("day_30_orders_dirty.csv")
df
```

## Exercise 1 — IQR Method

```python
# TODO

q1 = df["order_amount"].quantile(0.25)
q3 = df["order_amount"].quantile(0.75)
iqr = q3 - q1

lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

outlier_iqr = df[(df["order_amount"] < lower) | (df["order_amount"] > upper)]
outlier_iqr
```

## Exercise 2 — Z-score Method

```python
mean = df["quantity"].mean()
std = df["quantity"].std()

df["z_score"] = (df["quantity"] - mean) / std

outliers_z = df[df["z_score"].abs() > 3]
outliers_z
```

## Exercise 3 — Handle Outliers

```python
# TODO

df["order_amount"] = df["order_amount"].clip(lower, upper)
df
```

```python
# TODO

df["city"] = df["city"].astype("category")

df["order_id"] = pd.to_numeric(df["order_id"], downcast = "integer")
df["quantity"] = pd.to_numeric(df["quantity"], downcast = "integer")

df
```
