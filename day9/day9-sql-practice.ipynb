```python
import pandas as pd, sqlite3
customers = pd.read_csv('/kaggle/input/day9-datasets/day_9_customers.csv')
orders = pd.read_csv('/kaggle/input/day9-datasets/day_9_orders.csv')
conn = sqlite3.connect(':memory:')
customers.to_sql('customers', conn, index=False, if_exists='replace')
orders.to_sql('orders', conn, index=False, if_exists='replace')
```

**Show all customers who live in 'Mumbai'**

```python
query = """
SELECT *
FROM customers
WHERE city = 'Mumbai';
"""

pd.read_sql(query, conn)
```

**List customers aged between 22 and 26 (inclusive).**

```python
query = """
SELECT *
FROM customers
WHERE age BETWEEN 22 and 26;
"""

pd.read_sql(query, conn)
```

**Show orders with amount greater than 1000.**

```python
query = """
SELECT *
FROM orders
WHERE amount > 1000;
"""

pd.read_sql(query, conn)
```

**Display top 5 orders by amount (highest first).**

```python
query = """
SELECT *
FROM orders
ORDER BY amount desc
LIMIT 5 ;
"""

pd.read_sql(query, conn)
```

**List the 3 youngest customers (age ascending).**

```python
query = """
SELECT *
FROM customers
ORDER BY age asc
LIMIT 3 ;
"""

pd.read_sql(query, conn)
```

**Show customers whose name contains 'an' (case-insensitive - use LIKE).**

```python
query = """
SELECT *
FROM customers
WHERE name LIKE '%an%';
"""

pd.read_sql(query, conn)
```

**Show customers located in either 'Delhi', 'Mumbai', or 'Pune' (use IN)**

```python
query = """
SELECT *
FROM customers
WHERE city In ( 'Delhi','Mumbai', 'Pune');
"""

pd.read_sql(query, conn)
```

**Calculate total revenue across all orders.**

```python
query = """
SELECT SUM(amount)
FROM orders
"""

pd.read_sql(query, conn)
```

**Find the highest and lowest order amounts.**

```python
query = """
SELECT MAX(amount)
FROM orders
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT MIN(amount)
FROM orders
"""

pd.read_sql(query, conn)
```

**Revenue per category (category, SUM(amount)).**

```python
query = """
SELECT category, SUM(amount)
FROM orders
GROUP BY category;
"""

pd.read_sql(query, conn)
```

**Revenue per customer (customer_id, SUM(amount)).**

```python
query = """
SELECT customer_id, SUM(amount)
FROM orders
GROUP BY customer_id;
"""

pd.read_sql(query, conn)
```

**Show categories with total revenue > 3000 (use HAVING).**

```python
query = """
SELECT category, SUM(amount) as total_revenue
FROM orders
GROUP BY category
HAVING SUM(amount) > 3000;
"""

pd.read_sql(query, conn)
```

**Find customers whose age is greater than the average customer age.**

```python
query = """
SELECT *
FROM customers
WHERE age > (
    SELECT AVG(age) from customers
);
"""

pd.read_sql(query, conn)
```

**List customers who have placed at least one order.**

```python
query = """
SELECT *
FROM customers
WHERE customer_id IN (
    SELECT DISTINCT customer_id
    From orders
);
"""

pd.read_sql(query, conn)
```

**List customers who have never placed an order.**

```python
query = """
SELECT *
FROM customers
WHERE customer_id NOT IN (
    SELECT DISTINCT customer_id
    From orders
);
"""

pd.read_sql(query, conn)
```

**Find orders with amount greater than the average order amount.**

```python
query = """
SELECT *
FROM orders
WHERE amount > (
    SELECT AVG(amount)
    FROM orders
);
"""

pd.read_sql(query, conn)
```

**Find the second highest order amount.**

```python
query = """
SELECT MAX(amount) as second_highest
FROM orders
WHERE amount < (
    SELECT MAX(amount)
    FROM orders
);
"""

pd.read_sql(query, conn)
```

**Find top 2 customers by total spending (customer_id, total_spent).**

```python
query = """
SELECT customer_id, SUM(amount) as total_spent
FROM orders
GROUP BY customer_id
ORDER BY amount DESC
LIMIT 2;
"""

pd.read_sql(query, conn)
```
