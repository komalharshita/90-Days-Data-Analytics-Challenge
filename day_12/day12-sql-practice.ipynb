```python
import pandas as pd
import sqlite3

df = pd.read_csv('/kaggle/input/day-12-hotel-bookings/day_12_hotel_bookings.csv')
conn = sqlite3.connect(':memory:')
df.to_sql('bookings', conn, index=False, if_exists='replace')
```

```python
pd.read_sql("SELECT * FROM bookings LIMIT 5;", conn)
```

**All bookings from Delhi**

```python
query = """
SELECT *
FROM bookings
WHERE city = 'Delhi';
"""

pd.read_sql(query, conn)
```

**All Suite room bookings**

```python
query = """
SELECT *
FROM bookings
WHERE room_type = 'Suite';
"""

pd.read_sql(query, conn)
```

**Bookings where total_amount > 15000**

```python
query = """
SELECT *
FROM bookings
WHERE total_amount > 15000;
"""

pd.read_sql(query, conn)
```

**Top 5 most expensive bookings**

```python
query = """
SELECT *
FROM bookings
ORDER BY total_amount DESC
LIMIT 5 ;
"""

pd.read_sql(query, conn)
```

**Cheapest 5 bookings**

```python
query = """
SELECT *
FROM bookings
ORDER BY total_amount ASC
LIMIT 5 ;
"""

pd.read_sql(query, conn)
```

**Top 3 longest stays**

```python
query = """
SELECT *
FROM bookings
ORDER BY nights DESC
LIMIT 3 ;
"""

pd.read_sql(query, conn)
```

**Hotels containing 'Taj'**

```python
query = """
SELECT *
FROM bookings
WHERE hotel_name LIKE '%Taj%' ;
"""

pd.read_sql(query, conn)
```

**Customer names starting with 'K'**

```python
query = """
SELECT *
FROM bookings
WHERE customer_name LIKE 'K%' ;
"""

pd.read_sql(query, conn)
```

**Cities IN ('Delhi','Mumbai','Goa')**

```python
query = """
SELECT *
FROM bookings
WHERE city IN ('Delhi', 'Mumbai', 'Goa') ;
"""

pd.read_sql(query, conn)
```

**Total revenue**

```python
query = """
SELECT SUM(total_amount) as Total_Revenue
FROM bookings ;
"""

pd.read_sql(query, conn)
```

**Average booking value**

```python
query = """
SELECT AVG(total_amount) as Average_amount
FROM bookings
;
"""

pd.read_sql(query, conn)
```

**Total bookings count**

```python
query = """
SELECT COUNT(*) as Bookings_count
FROM bookings
 ;
"""

pd.read_sql(query, conn)
```

> monthly revenue

```python
query = """
SELECT booking_month, SUM(total_amount) AS monthly_revenue
FROM bookings
GROUP BY booking_month
ORDER BY monthly_revenue DESC;
"""

pd.read_sql(query, conn)
```

> revenue per region

```python
query = """
SELECT region, SUM(total_amount) AS revenue
FROM bookings
GROUP BY region
ORDER BY revenue DESC;
"""

pd.read_sql(query, conn)
```

> monthly revenue per region

```python
query = """
SELECT region, booking_month, SUM(total_amount) AS revenue
FROM bookings
GROUP BY region, booking_month
ORDER BY region, booking_month;
"""

pd.read_sql(query, conn)
```

> -- filter groups example

```python
query = """
SELECT region, booking_month, SUM(total_amount) AS revenue
FROM bookings
GROUP BY region, booking_month
HAVING SUM(total_amount) > 50000
ORDER BY revenue DESC; 
"""

pd.read_sql(query, conn)
```
