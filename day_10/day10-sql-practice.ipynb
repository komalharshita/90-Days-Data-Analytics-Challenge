```python
import pandas as pd
import sqlite3

movies = pd.read_csv('/kaggle/input/day-10-movies/day_10_movies.csv')

conn = sqlite3.connect(':memory:')
movies.to_sql('movies', conn, index=False, if_exists='replace')
```

```python
pd.read_sql("SELECT * FROM movies LIMIT 5;", conn)
```

# SELECT & WHERE:

**Show all movies released after 2018**

```python
query = """
SELECT *
FROM movies
WHERE release_year >= 2018;
"""

pd.read_sql(query, conn)
```

**Show movies where rating > 8**

```python
query = """
SELECT *
FROM movies
WHERE rating > 8;
"""

pd.read_sql(query, conn)
```

**Show movies in language = 'Korean'**

```python
query = """
SELECT *
FROM movies
WHERE language = "Korean";
"""

pd.read_sql(query, conn)
```

# ORDER BY & LIMIT:

**Top 5 highest rated movies**

```python
query = """
SELECT *
FROM movies
ORDER BY rating DESC
LIMIT 5;
"""

pd.read_sql(query, conn)
```

**Top 3 movies with highest views**

```python
query = """
SELECT *
FROM movies
ORDER BY views DESC
LIMIT 3;
"""

pd.read_sql(query, conn)
```

**Oldest 5 movies**

```python
query = """
SELECT *
FROM movies
ORDER BY release_year ASC
LIMIT 5 ;
"""

pd.read_sql(query, conn)
```

# LIKE & IN:

**Movies where title contains 'The'**

```python
query = """
SELECT *
FROM movies
WHERE title LIKE '%The%';
"""

pd.read_sql(query, conn)
```

**Movies where genre IN ('Drama', 'Romance')**

```python
query = """
SELECT *
FROM movies
WHERE genre in ('Drama', 'Romance');
"""

pd.read_sql(query, conn)
```

**Movies where language IN ('Hindi', 'Korean')**

```python
query = """
SELECT *
FROM movies
WHERE language IN ('Hindi', 'Korean');
"""

pd.read_sql(query, conn)
```

# Aggregates:

**Total number of movies**

```python
query = """
SELECT COUNT(*) as No_of_movies
FROM movies;
"""

pd.read_sql(query, conn)
```

**Average movie rating**

```python
query = """
SELECT AVG(rating) as Average_Rating
FROM movies
;
"""

pd.read_sql(query, conn)
```

**Total views across all movies**

```python
query = """
SELECT SUM(views) as Total_views
FROM movies
;
"""

pd.read_sql(query, conn)
```

# GROUP BY & HAVING:

**Number of movies per genre**

```python
query = """
SELECT genre, COUNT(*) as no_movies
FROM movies
GROUP BY genre;
"""

pd.read_sql(query, conn)
```

**Average rating per language**

```python
query = """
SELECT language, AVG(rating)
FROM movies
GROUP BY language;
"""

pd.read_sql(query, conn)
```

**Countries with average rating > 8**

```python
query = """
SELECT AVG(rating), country
FROM movies
GROUP BY Country
HAVING AVG(rating) > 8;
"""

pd.read_sql(query, conn)
```
