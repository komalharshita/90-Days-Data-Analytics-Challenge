```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

# UPDATE PATH BELOW
df = pd.read_csv("/kaggle/input/datasets/satyajeetrai/palmer-penguins-dataset-for-eda/penguins.csv")
df.head()
```

## Basic Cleaning

```python

numeric_cols = [
    "bill_length_mm",
    "bill_depth_mm",
    "flipper_length_mm",
    "body_mass_g",
    "year"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.dropna(subset=numeric_cols)
df.info()
```

## Scatterplot: Bill Length vs Bill Depth

```python

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species")
plt.title("Bill Length vs Bill Depth")
plt.show()
```

## Scatterplot with Trend Line

```python

plt.figure(figsize=(8,6))
sns.regplot(data=df, x="bill_length_mm", y="body_mass_g")
plt.title("Bill Length vs Body Mass (Trend Line)")
plt.show()
```

## Scatterplot: Flipper Length vs Body Mass

```python

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.title("Flipper Length vs Body Mass")
plt.show()
```
