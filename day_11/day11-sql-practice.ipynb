```python
import pandas as pd
import sqlite3

df = pd.read_csv('/kaggle/input/day11-dataset/day_11_clothing_accessories.csv')

conn = sqlite3.connect(':memory:')
df.to_sql('products', conn, index=False, if_exists='replace')

pd.read_sql("SELECT * FROM products LIMIT 5;", conn)
```

**Show all products where category = 'Clothing'**

```python
query = """
SELECT *
FROM products
WHERE category = 'Clothing';
"""

pd.read_sql(query, conn)
```

**Show all products priced above 2000**

```python
query = """
SELECT *
FROM products
WHERE price > 2000;
"""

pd.read_sql(query, conn)
```

**Show products targeted for 'Women'**

```python
query = """
SELECT *
FROM products
WHERE target_gender = 'Women';
"""

pd.read_sql(query, conn)
```

**Show products sold in city = 'Mumbai'**

```python
query = """
SELECT *
FROM products
WHERE city = 'Mumbai';
"""

pd.read_sql(query, conn)
```

**Show products with rating >= 4.5**

```python
query = """
SELECT *
FROM products
WHERE rating >= 4.5 ;
"""

pd.read_sql(query, conn)
```

**Show products where price is between 500 and 1500**

```python
query = """
SELECT *
FROM products
WHERE price BETWEEN 500 AND 1500 ;
"""

pd.read_sql(query, conn)
```

**Show products where category = 'Footwear' and price > 2500**

```python
query = """
SELECT *
FROM products
WHERE category = 'Footwear' AND price >= 2500 ;
"""

pd.read_sql(query, conn)
```

**Show products sorted by price (highest first)**

```python
query = """
SELECT *
FROM products
ORDER BY price DESC ;
"""

pd.read_sql(query, conn)
```

**Show products where city is either Delhi or Bangalore**

```python
query = """
SELECT *
FROM products
WHERE city = 'Delhi' OR city = 'Bangalore';
"""

pd.read_sql(query, conn)
```

**Show top 5 most expensive products**

```python
query = """
SELECT *
FROM products
ORDER BY price DESC
LIMIT 5 ;
"""

pd.read_sql(query, conn)
```

**Find products whose rating is higher than average rating**

```python
query = """
SELECT *
FROM products
WHERE rating > (
    SELECT AVG(rating) from products
);
"""

pd.read_sql(query, conn)
```

**Find products whose price is greater than the average price**

```python
query = """
SELECT *
FROM products
WHERE price > (
    SELECT AVG(price) FROM products
);
"""

pd.read_sql(query, conn)
```

**Find products belonging to the most expensive category**

```python
query = """
SELECT * FROM products
WHERE category = (
    SELECT category
    FROM Products
    GROUP BY category
    ORDER BY AVG(price) DESC
    LIMIT 1
) ;
"""

pd.read_sql(query, conn)
```

**Find brands that have at least one product priced above 3000**

```python
query = """
SELECT DISTINCT brand
FROM products
WHERE brand IN (
    SELECT brand
    FROM products
    WHERE price > 3000
);
"""

pd.read_sql(query, conn)
```

**Create a derived table showing total units_sold per category**

```python
query = """
SELECT *
FROM (
    SELECT category, SUM(units_sold) as units
    FROM products
    GROUP BY category
) AS table1 ;
"""

pd.read_sql(query, conn)
```

**From that derived table, display only categories where total units_sold > 1000**

```python
query = """
SELECT *
FROM (
    SELECT category, SUM(units_sold) AS total_units
    FROM products
    GROUP BY category
) AS table2
WHERE total_units > 1000;

"""

pd.read_sql(query, conn)
```

**Create a derived table showing average price per brand**

```python
query = """
SELECT *
FROM (
    SELECT brand, AVG(price) AS average_price
    FROM products
    GROUP BY brand
) AS table3 ;

"""

pd.read_sql(query, conn)
```

**From that derived table, sort brands by average price**

```python
query = """
SELECT *
FROM (
    SELECT brand, AVG(price) AS avg_price
    FROM products
    GROUP BY brand
    ORDER BY AVG(price) DESC 
) AS table4 ;

"""

pd.read_sql(query, conn)
```
