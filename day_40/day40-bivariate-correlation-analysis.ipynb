```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv("/kaggle/input/datasets/joebeachcapital/linkedin-jobs/final_data.csv", encoding="latin1")

df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
numeric_cols = [
    "Total_applicants",
    "LinkedIn_Followers",
    "Employee_count"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.dropna(subset=numeric_cols)

df.info()
```

# Scatterplot Between Two Important Metrics - Followers vs Applicants

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,
                x="LinkedIn_Followers",
                y="Total_applicants")

plt.title("Followers vs Total Applicants")
plt.show()
```

# Add regression trend line

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df,
            x="LinkedIn_Followers",
            y="Total_applicants",
            scatter_kws={"alpha":0.3})

plt.title("Trend: Followers vs Applicants")
plt.show()
```
