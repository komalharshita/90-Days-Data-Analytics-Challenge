# Day 33 — Date Parsing & String Cleaning (Practice)

```python

import pandas as pd
df = pd.read_csv("day_33_event_registrations_dirty.csv")
df
```

## Exercise 1 — Parse registration_date

```python
# TODO

df["registration_date"] = pd.to_datetime(
    df["registration_date"], errors = "coerce"
)
```

## Exercise 2 — Identify invalid dates

```python
# TODO

invalid_dates = df[df["registration_date"].isna()]
invalid_dates
```

## Exercise 3 — Extract date parts

```python
# TODO

df["year"] = df["registration_date"].dt.year
df["month"] = df["registration_date"].dt.month
df["day"] = df["registration_date"].dt.day
```

## Exercise 4 — Clean event_name

```python
# TODO

df["event_name_clean"] = (
    df["event_name"]
    .str.strip()
    .str.lower()
    .str.replace(r"[^a-z\s-]", "", regex = True)
    .str.replace(r"\s+", "", regex= True)
)
```

## Exercise 5 — Clean city names

```python
# TODO

df["city_clean"] = (
    df["city"]
    .str.strip()
    .str.lower()
    .str.replace(r"[^a-z]", "", regex = True)
)

df
```
