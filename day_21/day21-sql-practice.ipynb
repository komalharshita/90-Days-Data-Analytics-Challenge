# ðŸŸ¢ EASY LEVEL (Foundations + JOIN + GROUP BY)

**Show all customers and their cities.**

```python
pd.read_sql("""
SELECT name, city
FROM customers
; 
""", conn)
```

**Show all purchases where amount > 700.**

```python
pd.read_sql("""
SELECT *
FROM purchases
WHERE amount > 700
; 
""", conn)
```

**Show customer names with their purchased product (INNER JOIN).**

```python
pd.read_sql("""
SELECT name, product
FROM customers c
INNER JOIN purchases p
ON c.customer_id = p.customer_id 
; 
""", conn)
```

**Show total spending per customer.**

```python
pd.read_sql("""
SELECT name, SUM(amount) AS total_spending 
FROM customers c
INNER JOIN purchases p
ON c.customer_id = p.customer_id 
GROUP BY name
; 
""", conn)
```

**Show number of purchases per category.**

```python
pd.read_sql("""
SELECT category, COUNT(*) AS total_purchases
FROM customers c
INNER JOIN purchases p
ON c.customer_id = p.customer_id 
GROUP BY category
; 
""", conn)
```

**Show total revenue per city.**

```python
pd.read_sql("""
SELECT city, SUM(amount) as total_revenue
FROM customers c
INNER JOIN purchases p
ON c.customer_id = p.customer_id 
GROUP BY city
; 
""", conn)
```

**Show customers who belong to the "Curly" hair_type.**

# ðŸŸ¡ MEDIUM LEVEL (CASE, Subqueries, Window Basics, Aggregation Logic)

**Label each purchase as:
"Expensive" if amount â‰¥ 900
"Affordable" otherwise**

```python
pd.read_sql("""
SELECT purchase_id, product, amount, CASE
    WHEN amount >= 900 THEN "Expensive"
    ELSE "Affordable"
    END AS purchase_type
FROM purchases;
 
""", conn)
```

**Show customers whose total spending is greater than average customer spending.**

```python
pd.read_sql("""
SELECT customer_id, SUM(amount) AS total_spent
FROM purchases
GROUP BY customer_id
HAVING SUM(amount) >
(
  SELECT AVG(total)
  FROM (
    SELECT SUM(amount) AS total
    FROM purchases
    GROUP BY customer_id
  )
);

""", conn)
```

**Show category-wise revenue and classify:
"Top Category" if revenue â‰¥ overall average revenue
"Low Category" otherwise**

```python
pd.read_sql("""
SELECT category, total_revenue,
CASE
  WHEN total_revenue >= avg_rev THEN 'Top Category'
  ELSE 'Low Category'
END AS label
FROM (
  SELECT category,
         SUM(amount) AS total_revenue,
         AVG(SUM(amount)) OVER () AS avg_rev
  FROM purchases
  GROUP BY category
);

""", conn)
```

**Show each purchase along with the customer's city.**

```python
pd.read_sql("""
SELECT p.*, c.city
FROM purchases p
JOIN customers c
ON p.customer_id = c.customer_id;

 
""", conn)
```

**Customers with more than one purchase**

```python
pd.read_sql("""
SELECT customer_id, product
FROM purchases
GROUP BY customer_id
HAVING COUNT(*) > 1;
 
""", conn)
```

**Show top 3 most expensive purchases.**

```python
pd.read_sql("""
SELECT *
FROM purchases
ORDER BY amount DESC
LIMIT 3;
 
""", conn)
```

**For each city, show total revenue.**

```python
pd.read_sql("""
SELECT c.city, SUM(p.amount) AS revenue
FROM customers c
JOIN purchases p
ON c.customer_id = p.customer_id
GROUP BY c.city;

""", conn)
```

**First purchase date per customer**

```python
pd.read_sql("""
SELECT customer_id, MIN(purchase_date) AS first_purchase
FROM purchases
GROUP BY customer_id;
 
""", conn)
```

**Show purchases that are higher than the customer's own average purchase amount.**

```python
pd.read_sql("""
SELECT *
FROM (
  SELECT *,
         AVG(amount) OVER (PARTITION BY customer_id) AS avg_amt
  FROM purchases
)
WHERE amount > avg_amt;

""", conn)
```

**Show percentage contribution of each category to total revenue.**

```python
pd.read_sql("""
SELECT category,
       SUM(amount) * 100.0 / SUM(SUM(amount)) OVER () AS percent_contribution
FROM purchases
GROUP BY category;

""", conn)
```

# ðŸ”´ HARD LEVEL (CTEs, Window Functions, Lead/Lag, Multi-step Logic)

**Rank customers by total spending (highest spender = rank 1).**

```python
pd.read_sql("""
SELECT customer_id, total_spent,
       RANK() OVER (ORDER BY total_spent DESC) AS rank
FROM (
  SELECT customer_id, SUM(amount) AS total_spent
  FROM purchases
  GROUP BY customer_id
);
 
""", conn)
```

**Rank customers by total spending (highest spender = rank 1).**

```python
pd.read_sql("""
SELECT *
FROM (
  SELECT c.city, c.name,
         SUM(p.amount) AS total_spent,
         RANK() OVER (
           PARTITION BY c.city
           ORDER BY SUM(p.amount) DESC
         ) AS rnk
  FROM customers c
  JOIN purchases p
  ON c.customer_id = p.customer_id
  GROUP BY c.city, c.name
)
WHERE rnk <= 2;

 
""", conn)
```

**Show running total spending per customer ordered by purchase_date.**

```python
pd.read_sql("""
SELECT customer_id, purchase_date, amount,
       SUM(amount) OVER (
         PARTITION BY customer_id
         ORDER BY purchase_date
       ) AS running_total
FROM purchases;
 
""", conn)
```

**Show running total spending per customer.**

```python
pd.read_sql("""
SELECT customer_id, purchase_date, amount,
       SUM(amount) OVER (
         PARTITION BY customer_id
         ORDER BY purchase_date
       ) AS running_total
FROM purchases;
 
""", conn)
```

**For each purchase, show current amount and previous purchase amount for that customer (use LAG).**

```python
pd.read_sql("""
SELECT customer_id, purchase_date, amount,
       LAG(amount) OVER (
         PARTITION BY customer_id
         ORDER BY purchase_date
       ) AS previous_amount
FROM purchases;

""", conn)
```

**For each category, calculate month-wise revenue (use SUBSTR on purchase_date).**

```python
pd.read_sql("""
WITH totals AS (
  SELECT customer_id, SUM(amount) AS total_spent
  FROM purchases
  GROUP BY customer_id
)
SELECT *
FROM totals
WHERE total_spent > 1500;

""", conn)
```

**Show customers whose spending is above their cityâ€™s average spending.**

```python
pd.read_sql("""
SELECT *
FROM (
  SELECT c.customer_id, c.city,
         SUM(p.amount) AS total_spent,
         AVG(SUM(p.amount)) OVER (PARTITION BY c.city) AS city_avg
  FROM customers c
  JOIN purchases p
  ON c.customer_id = p.customer_id
  GROUP BY c.customer_id, c.city
)
WHERE total_spent > city_avg;
 
""", conn)
```

**Most purchased product per category**

```python
pd.read_sql("""
SELECT category, product
FROM (
  SELECT category, product,
         COUNT(*) AS cnt,
         RANK() OVER (
           PARTITION BY category
           ORDER BY COUNT(*) DESC
         ) AS rnk
  FROM purchases
  GROUP BY category, product
)
WHERE rnk = 1;

""", conn)
```

```python
pd.read_sql("""
SELECT c.acquisition_channel,
       SUM(p.amount) * 100.0 / SUM(SUM(p.amount)) OVER () AS percent_contribution
FROM customers c
JOIN purchases p
ON c.customer_id = p.customer_id
GROUP BY c.acquisition_channel;

""", conn)
```

**Top 3 customers by lifetime value (CTE + window)**

```python
pd.read_sql("""
WITH totals AS (
  SELECT customer_id, SUM(amount) AS total_spent
  FROM purchases
  GROUP BY customer_id
),
ranked AS (
  SELECT customer_id, total_spent,
         RANK() OVER (ORDER BY total_spent DESC) AS rnk
  FROM totals
)
SELECT *
FROM ranked
WHERE rnk <= 3;

""", conn)
```
