## Kaggle Setup

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

# UPDATE PATH
df = pd.read_csv("/kaggle/input/datasets/satyajeetrai/palmer-penguins-dataset-for-eda/penguins.csv")
df.head()
```

## 1. Context
Describe the question you are exploring.

3. does bill length correlate with body mass?

## 2. Exploratory Overview

```python

df.describe()
```

## 3. Species Comparison

```python

species_summary = df.groupby("species")["body_mass_g"].mean()
species_summary
```

```python

sns.barplot(x=species_summary.index, y=species_summary.values)
plt.title("Average Body Mass by Species")
plt.show()
```

## 4. Relationship Analysis

```python

sns.scatterplot(data=df, x="bill_length_mm", y="body_mass_g", hue="species")
plt.title("Bill Length vs Body Mass")
plt.show()
```
