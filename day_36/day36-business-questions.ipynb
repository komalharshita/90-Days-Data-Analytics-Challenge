```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv("/kaggle/input/datasets/mohammadtalib786/retail-sales-dataset/retail_sales_dataset.csv")
df.head()
```

**This dataset contains retail transaction data including sales,
profit, quantity, category, and region.
The goal is to understand basic sales behavior.**


# Question 1 — Are sales values evenly distributed?

```python
sns.histplot(df["Total Amount"], bins=30, kde=True)
plt.title("Are Sales Evenly Distributed?")
plt.show()
```

# Question 2 — Are there extreme sales values?

```python
sns.boxplot(x=df["Total Amount"])
plt.title("Do Sales Have Outliers?")
plt.show()
```

# Question 3 — What is the category distribution?

```python
sns.histplot(df["Product Category"], bins=30, kde=True)
plt.title("Category Distribution")
plt.show()
```

# Question 4 — Are most orders small or large?

```python
sns.histplot(df["Quantity"], bins=20)
plt.title("Order Size Distribution")
plt.show()
```
