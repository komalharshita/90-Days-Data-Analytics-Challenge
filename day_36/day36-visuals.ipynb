```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/datasets/mohammadtalib786/retail-sales-dataset/retail_sales_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.histplot(df["Total Amount"], bins = 30, kde = True)
plt.title("Distribution of Sales")
plt.show()
```

```python
sns.boxplot(x=df["Total Amount"])
plt.title("Boxplot of Sales")
plt.show()
```

```python
sns.histplot(df["Quantity"], bins = 20, kde = True)
plt.title("Distribution of Quantity sold")
plt.show()
```

```python
sns.boxplot(x=df["Age"])
plt.title("Boxplot of Age")
plt.show()
```
