## Exercise 1 — Merge orders with customer names

```python

# EXERCISE: Print customer name with item and cleaned price

with open("day_26_customers.csv", "r") as f:
    next(f)
    for line in f:
        parts = line.strip().split(",")
        customers[parts[0]] = parts[1]

with open("day_26_orders.csv", "r") as f:
    next(f)
    for line in f:
        parts = line.strip().split(",")
        cid = parts[1]
        item = parts[2]
        price = int(parts[3].replace("Rs.", ""))

        name = customers.get(cid, "UNKNOWN")
        print(name, item, price)
```

## Exercise 2 — Write merged output to CSV

```python

# EXERCISE: Create merged_output.csv with columns: name,item,price

with open("merged_output.csv", "w") as out:
    out.write("name, item, price\n")

    with open("day_26_orders.csv", "r") as f:
        next(f)
        for line in f:
            parts = line.strip().split(",")
            cid = parts[1]
            item = parts[2]
            price = parts[3].replace("Rs.", "")

            name = customers.get(cid, "UNKNOWN")
            out.write(f"{name}, {item}, {price}\n")
```
