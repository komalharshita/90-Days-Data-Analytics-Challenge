```python
import pandas as pd

customers = pd.read_csv('/kaggle/input/datasets-day8/day8_customers.csv')
orders = pd.read_csv('/kaggle/input/datasets-day8/day8_orders.csv')

customers.head()
```

```python
orders.head()
```

```python
import sqlite3

conn = sqlite3.connect(':memory:')
customers.to_sql('customers', conn, index=False, if_exists='replace')
orders.to_sql('orders', conn, index=False, if_exists='replace')
```

```python
query = """
SELECT *
FROM customers;
"""

pd.read_sql(query, conn)
```

**All customers from Delhi**

```python
query = """
Select * from customers
where city = 'Delhi'
"""

pd.read_sql(query, conn)
```

**Customers older than 22**

```python
query = """
SELECT * from customers
where age > 22
"""

pd.read_sql(query, conn)
```

**All orders above 1000**

```python
query = """
SELECT * from orders
where amount > 1000
"""

pd.read_sql(query, conn)
```

**Top 3 highest order amounts**

```python
query = """
SELECT * from orders
order by amount desc
limit 3 ;
"""

pd.read_sql(query, conn)
```

**Youngest two customers**

```python
query = """
SELECT * from customers
order by age asc
limit 2
"""

pd.read_sql(query, conn)
```

**Customers whose name starts with 'A'**

```python
query = """
SELECT * from customers
where name LIKE 'A%' ;
"""

pd.read_sql(query, conn)
```

**Customers from Delhi or Mumbai**

```python
query = """
SELECT * from customers
WHERE city IN ('Delhi', 'Mumbai')
"""

pd.read_sql(query, conn)
```

**Categories containing 'Elect'**

```python
query = """
SELECT * from orders
WHERE category LIKE '%Elect%'
"""

pd.read_sql(query, conn)
```

**Total revenue**

```python
query = """
SELECT SUM(amount)
from orders
"""

pd.read_sql(query, conn)
```

**Average order amount**

```python
query = """
SELECT AVG(amount)
from orders
"""

pd.read_sql(query, conn)
```

**Total number of orders**

```python
query = """
SELECT COUNT(*)
from orders
"""

pd.read_sql(query, conn)
```
