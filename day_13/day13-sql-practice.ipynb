```python
import pandas as pd
import sqlite3

df = pd.read_csv('/kaggle/input/day-13-online-courses/day_13_online_courses.csv')
conn = sqlite3.connect(':memory:')
df.to_sql('enrollments', conn, index=False, if_exists='replace')

pd.read_sql("SELECT * FROM enrollments LIMIT 10;", conn)
```

**All beginner level courses**

```python
query = """
SELECT *
FROM enrollments
WHERE level = 'Beginner';
"""

pd.read_sql(query, conn)
```

**All enrollments in category = 'Analytics'**

```python
query = """
SELECT *
FROM enrollments
WHERE category = 'Analytics';
"""

pd.read_sql(query, conn)
```

**Courses priced above 2000**

```python
query = """
SELECT course_name, course_price
FROM enrollments
WHERE course_price > 2000;
"""

pd.read_sql(query, conn)
```

**Top 5 most expensive courses**

```python
query = """
SELECT course_name, course_price
FROM enrollments
ORDER BY course_price DESC
LIMIT 5 ;
"""

pd.read_sql(query, conn)
```

**Lowest 5 rated courses**

```python
query = """
SELECT course_name, course_price, rating
FROM enrollments
ORDER BY rating ASC
LIMIT 5 ;
"""

pd.read_sql(query, conn)
```

**Top 3 highest rated courses**

```python
query = """
SELECT course_name, course_price, rating
FROM enrollments
ORDER BY rating DESC
LIMIT 3 ;
"""

pd.read_sql(query, conn)
```

**Course names containing 'SQL'**

```python
query = """
SELECT course_name
FROM enrollments
WHERE course_name LIKE '%SQL%' ;
"""

pd.read_sql(query, conn)
```

**Student names starting with 'A'**

```python
query = """
SELECT course_name, student_name, enrollment_id
FROM enrollments
WHERE student_name LIKE 'A%' ;
"""

pd.read_sql(query, conn)
```

**Categories IN ('Analytics','Data Science')**

```python
query = """
SELECT *
FROM enrollments
WHERE category IN ('Analytics', 'Data Science') ;
"""

pd.read_sql(query, conn)
```

**Total revenue (SUM of course_price)**

```python
query = """
SELECT SUM(course_price) AS Sum_of_course_price
FROM enrollments ;
"""

pd.read_sql(query, conn)
```

**Average course price**

```python
query = """
SELECT AVG(course_price) AS Avg_of_course_price
FROM enrollments ;
"""

pd.read_sql(query, conn)
```

**Total enrollments count**

```python
query = """
SELECT COUNT(*) AS TotalEnrollments
FROM enrollments ;
"""

pd.read_sql(query, conn)
```

**Revenue per category**

```python
query = """
SELECT category, SUM(course_price) AS Sum_of_course_price
FROM enrollments
GROUP BY category ;
"""

pd.read_sql(query, conn)
```

**Average rating per level**

```python
query = """
SELECT level, AVG(rating) AS AverageRating
FROM enrollments
GROUP BY level;
"""

pd.read_sql(query, conn)
```

**Categories where total revenue > 10000**

```python
query = """
SELECT category, SUM(course_price) AS TotalRevenue
FROM enrollments
GROUP BY category
HAVING SUM(course_price) > 10000 ;
"""

pd.read_sql(query, conn)
```

**Levels where average price > 1500**

```python
query = """
SELECT level, AVG(course_price) AS Average_price
FROM enrollments
GROUP BY level
HAVING AVG(course_price) > 1500 ;
"""

pd.read_sql(query, conn)
```
