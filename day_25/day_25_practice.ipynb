# Day 25 — Strings, Loops, Functions Practice

```python

rows = []
with open("day_25_cafe_orders.csv", "r") as f:
    next(f)
    for line in f:
        rows.append(line.strip())

rows[:3]
```

## Exercise 1 — Clean price and extract integer

```python

# EXERCISE: Clean price column using string methods and convert to int

clean_prices = []
for row in rows:
    parts = row.split(",")
    price = parts[4].replace("Rs.", "").strip()
    clean_prices.append(int(price))

print(clean_prices)
```

## Exercise 2 — Print customers from Delhi

```python

# EXERCISE: Use loop + condition to print names from Delhi

for row in rows:
    parts = row.split(",")
    if parts[2] == "Delhi":
        print(parts[1])
```

## Exercise 3 — Function to find average price

```python

# EXERCISE: Write a function average_price(rows)

def average_price(rows):
    total = 0
    count = 0
    for row in rows:
        parts = row.split(",")
        price = int(parts[4].replace("Rs.", "").strip())
        total += price 
        count += 1
    return total / count

average_price(rows)
```

## Exercise 4 — List of expensive orders (>250)

```python

# EXERCISE: Use list comprehension to find orders where price > 250

expensive = [
    row.split(",")[1]
    for row in rows
    if int(row.split(",")[4].replace("Rs.", "").strip()) > 250
]

print(expensive)
```
