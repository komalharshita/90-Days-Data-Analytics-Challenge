```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

df = pd.read_csv("/kaggle/input/datasets/thedrzee/top-selling-books-data2023-2025/best sellin books 2023.csv",
                 encoding="latin1")

df.head()
```

```python
df.info()
df.describe()
```

```python
df["id"] = df["id"].str.replace("#", "", regex=False)
df["id"] = pd.to_numeric(df["id"], errors="coerce")

df["Rating"] = df["Rating"].str.extract(r"(\d+\.\d+)")
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")

df["price"] = df["price"].str.replace("$", "", regex=False)
df["price"] = pd.to_numeric(df["price"], errors="coerce")

df["reviews count"] = pd.to_numeric(df["reviews count"], errors="coerce")

df["Publishing date"] = pd.to_datetime(
    df["Publishing date"],
    errors="coerce",
    dayfirst=True
)
df["Year"] = df["Publishing date"].dt.year

df["Genre"] = df["Genre"].str.strip()

df = df.drop(columns=["Reading age"])
```

```python
df.info()
```

# Genre Performance

```python
genre_stats = (
    df.groupby("Genre")["reviews count"]
    .mean()
    .sort_values(ascending= False)
)

genre_stats.head(10)
```

# Author dominance

```python
author_stats = (
    df.groupby("Author")["reviews count"]
    .mean()
    .sort_values(ascending = False)
)

author_stats.head(10)
```

# Format Impact (Hardcover vs Paperback vs Kindle)

```python
format_stats = (
    df.groupby("form")["reviews count"]
      .mean()
      .sort_values(ascending=False)
)

format_stats
```

# Price Impact on Popularity

```python
df["price"].corr(df["reviews count"])
```

# Time based analysis

```python
yearly_stats = (
    df.groupby("Year")["reviews count"]
      .mean()
)

yearly_stats
```

```python
sns.lineplot(x=yearly_stats.index,
             y=yearly_stats.values)
plt.title("Average Reviews by Year")
plt.show()
```
