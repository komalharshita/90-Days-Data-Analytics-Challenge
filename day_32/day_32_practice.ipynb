# Day 32 — Feature Engineering & Data Validation (Practice)

```python

import pandas as pd
import numpy as np
from datetime import datetime

df = pd.read_csv("day_32_logistics_shipments.csv")
df
```

## Exercise 1 — Parse datetime columns

```python
# TODO

df["dispatch_time"] = pd.to_datetime(df["dispatch_time"])
df["devlivery_time"] = pd.to_datetime(df["delivery_time"])
```

## Exercise 2 — Create delivery duration feature

```python
df["dispatch_time"] = pd.to_datetime(df["dispatch_time"])
df["delivery_time"] = pd.to_datetime(df["delivery_time"])
df.dtypes

df["delivery_duration_min"] = (
    df["delivery_time"] - df["dispatch_time"]
).dt.total_seconds() / 60
```

## Exercise 3 — Time-based features

```python
# TODO

df["dispatch_hour"] = df["dispatch_time"].dt.hour
df["dispatch_day"] = df["dispatch_time"].dt.day_name()
df["is_weekend"] = df["dispatch_time"].dt.weekday >= 5
```

## Exercise 4 — Numeric & bucket features

```python
# TODO

df["delivery_hours"] = df["delivery_duration_min"] / 60
df["speed_kmph"] = df["distance_km"] / df["delivery_hours"]

df["distance_bucket"] = pd.cut(
    df["distance_km"],
    bins=[0, 15, 30, np.inf],
    labels=["Short", "Medium", "Long"]
)
```

## Exercise 5 — Flag features

```python
# TODO

df["is_delayed"] = df["delivery_status"] == "Delayed"
df["is_night_delivery"] = df["dispatch_hour"] >= 20
```
