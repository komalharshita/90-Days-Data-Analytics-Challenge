## 4. Loops — for and while

Iteration examples and common patterns.

```python
# for loop example
nums = [10, 20, 30, 40]
sum_val = 0
for n in nums:
    sum_val += n
print('Sum:', sum_val)

# while loop example
count = 0
while count < 3:
    print('count', count)
    count += 1
```

## 5. Functions — define reusable code

Include docstrings and returns.

```python
def calculate_average(numbers):
    """Return the average of a list of numbers."""
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)

# Example usage
print(calculate_average([10, 20, 30]))
```

**EXERCISE:** Write a function `is_even(n)` that returns True if n is even, else False.

```python
# EXERCISE: write code here

def is_even(n):
    """function to calculate even or odd"""
    if n % 2 == 0:
        return True
    return False

print(is_even(4))  # True
```

**EXERCISE:** Given `people = [{'name':'A', 'age':20}, {'name':'B','age':25}]` write code to extract names into a list.

## 7. List Comprehensions

Short, pythonic transformations.

```python
# Example
nums = [1,2,3,4,5]
sq = [n*n for n in nums if n%2==1]
print(sq)
```
