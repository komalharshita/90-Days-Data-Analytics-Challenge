# Feature engineering

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"].fillna(0, inplace=True)
```

```python
df["Churn_Flag"] = df["Churn"].map({"Yes": 1, "No": 0})
```

```python
df["Tenure_Group"] = pd.cut(
    df["tenure"],
    bins=[-1, 12, 24, 48, 60, 100],
    labels=["0-1yr", "1-2yr", "2-4yr", "4-5yr", "5+yr"]
)
```

```python
df["Spend_Category"] = pd.cut(
    df["MonthlyCharges"],
    bins=[0, 30, 60, 90, 120],
    labels=["Low", "Medium", "High", "Very High"],
    include_lowest=True
)
```

```python
df["Customer_Value"] = df["TotalCharges"]
```

## Visual analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Churn analysis 

```python
sns.countplot(data = df, x= "Churn")
plt.show()
```

## Tenure group vs Churn

```python
df.groupby("Tenure_Group")["Churn_Flag"].mean().plot(kind = 'bar')
plt.title("Churn rate by Tenure group")
plt.show()
```

# Key insights

- month to month costomers show significantly higher churn rates
- customers with tenure less than 1 year are at highest risk
- high monthly charges correlate with increased churn
- long term contracts drastically reduce churn

# Business impact

By targeting high risk segments:
- reduce churn by focusing on new customers
- increase retention through long term plans
- improve revenue by retaining high value users
